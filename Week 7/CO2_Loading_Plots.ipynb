```julia
using Plots
using DataFrames
using CSV
using LaTeXStrings
using Clapeyron
```

```julia
exp_298 = CSV.read("MEA_H2O_CO2_loading_298.15.csv",
               header=[1, 2, 3],
               DataFrame);

exp_313 = CSV.read("MEA_H2O_CO2_loading_313.15.csv",
               header=[1, 2, 3],
               DataFrame);

exp_333 = CSV.read("MEA_H2O_CO2_loading_333.15.csv",
               header=[1, 2, 3],
               DataFrame);

exp_353 = CSV.read("MEA_H2O_CO2_loading_353.15.csv",
               header=[1, 2, 3],
               DataFrame);
    
exp_373 = CSV.read("MEA_H2O_CO2_loading_373.15.csv",
               header=[1, 2, 3],
               DataFrame);
```

```julia
plt = plot(grid = :off, 
            framestyle = :box, 
            foreground_color_legend = nothing, 
            legend_font = font(12),
            yaxis = :log,
            xtickfont = font(12),
            ytickfont = font(12),
            #xguidefontsize = font(12),
            xlims = (0, 1),
            #legend = :outertopright,
            #legend = :outerbottom,
            legend = :topleft,
            size = (600, 400),
            #left_margin = 10mm
            )

plot!(plt, exp_373[:, 2], exp_373[:, 1]./1e5,
            seriestype = :scatter, 
            color = "white", 
            markerstrokecolor = "indianred2",
            marker = :utriangle,
            markerstrokewidth = 1,
            line = (:scatter, 3), 
            label = "373.15K")

plot!(plt, exp_353[:, 2], exp_353[:, 1]./1e5,
            seriestype = :scatter, 
            color = "white", 
            markerstrokecolor = "springgreen4",
            marker = :square,
            markerstrokewidth = 1,
            line = (:scatter, 3), 
            label = "353.15K")
    
plot!(plt, exp_333[:, 2], exp_333[:, 1]./1e5,
            seriestype = :scatter, 
            color = "white", 
            markerstrokecolor = "deepskyblue1",
            marker = :circle,
            markerstrokewidth = 1,
            line = (:scatter, 3), 
            label = "333.15K")

plot!(plt, exp_313[:, 2], exp_313[:, 1]./1e5,
            seriestype = :scatter, 
            color = "white", 
            markerstrokecolor = "mediumorchid2",
            marker = :diamond,
            markerstrokewidth = 1,
            line = (:scatter, 3), 
            label = "313.15K")

plot!(plt, exp_298[:, 2], exp_298[:, 1]./1e5,
            seriestype = :scatter, 
            color = "white", 
            markerstrokecolor = "chocolate1",
            marker = :hexagon,
            markerstrokewidth = 1,
            line = (:scatter, 3), 
            label = "298.15K")

xlabel!(plt, ""*L"\mathrm{_{CO_2}}")
ylabel!(plt, "P"*L"\mathrm{_{CO_2}}"*" / Pa")
#ylabel!(plt, "test")
```
